## Parsing and Matching
Parsing (hmdb_metabolites.xml) file using ElementTree package, then match the output with df1 from FooDB

```python
#Use a suitable device with RAM more than 8GB for this!
tree = ET.parse('C:/Users/Sara/Desktop/git_101/HMDB/hmdb_metabolites.xml')
root= tree.getroot()
```

```python
child= root[0]
child.tag
for element in child:
    print(element.tag)
    print(element.text)
```

```python
cols = ['name', 'iupac_name', 'traditional_iupac','inchikey', 'inchi', 'chemical_formula',
        'average_molecular_weight', 'description', 'synonyms', 'normal_concentrations', 'abnormal_concentrations',
        'chemspider_id', 'drugbank_id', 'foodb_id', 'pubchem_compound_id', 'pdb_id', 'chebi_id',
        'kegg_id', 'biocyc_id', 'vmh_id', 'protein_associations'
       ]
rows = []

for metabolite in root:
    name = metabolite.find('{http://www.hmdb.ca}name')
    if name is not None: name = name.text
    
    iupac_name = metabolite.find('{http://www.hmdb.ca}iupac_name')
    if iupac_name is not None: iupac_name = iupac_name.text

    traditional_iupac = metabolite.find('{http://www.hmdb.ca}traditional_iupac')
    if traditional_iupac is not None: traditional_iupac = traditional_iupac.text

    inchikey = metabolite.find('{http://www.hmdb.ca}inchikey')
    if inchikey is not None: inchikey = inchikey.text
    
    inchi = metabolite.find('{http://www.hmdb.ca}inchi')
    if inchi is not None: inchi = inchi.text
    
    chemical_formula = metabolite.find('{http://www.hmdb.ca}chemical_formula')
    if chemical_formula is not None: chemical_formula = chemical_formula.text
    
    average_molecular_weight = metabolite.find('{http://www.hmdb.ca}average_molecular_weight')
    if average_molecular_weight is not None: average_molecular_weight = average_molecular_weight.text
    
    description = metabolite.find('{http://www.hmdb.ca}description')
    if description is not None: description = description.text
    
    synonyms = metabolite.find('{http://www.hmdb.ca}synonyms')
    if synonyms is not None: synonyms = synonyms.text
    
    normal_concentrations= metabolite.find('{http://www.hmdb.ca}normal_concentrations')
    if normal_concentrations is not None: normal_concentrations = normal_concentrations.text
    
    abnormal_concentrations= metabolite.find('{http://www.hmdb.ca}abnormal_concentrations')
    if abnormal_concentrations is not None: abnormal_concentrations = abnormal_concentrations.text
    
    chemspider_id= metabolite.find('{http://www.hmdb.ca}chemspider_id')
    if chemspider_id is not None: chemspider_id = chemspider_id.text
    
    drugbank_id= metabolite.find('{http://www.hmdb.ca}drugbank_id')
    if drugbank_id is not None: drugbank_id = drugbank_id.text
    
    foodb_id = metabolite.find('{http://www.hmdb.ca}foodb_id')
    if foodb_id is not None: foodb_id = foodb_id.text
    
    pubchem_compound_id = metabolite.find('{http://www.hmdb.ca}pubchem_compound_id')
    if pubchem_compound_id is not None: pubchem_compound_id = pubchem_compound_id.text
    
    pdb_id = metabolite.find('{http://www.hmdb.ca}pdb_id')
    if pdb_id is not None: pdb_id = pdb_id.text
    
    chebi_id = metabolite.find('{http://www.hmdb.ca}chebi_id')
    if chebi_id is not None: chebi_id = chebi_id.text
    
    kegg_id = metabolite.find('{http://www.hmdb.ca}kegg_id')
    if kegg_id is not None: kegg_id = kegg_id.text
    
    biocyc_id = metabolite.find('{http://www.hmdb.ca}biocyc_id')
    if biocyc_id is not None: biocyc_id = biocyc_id.text
    
    vmh_id = metabolite.find('{http://www.hmdb.ca}vmh_id')
    if vmh_id is not None: vmh_id = vmh_id.text
    
    protein_associations = metabolite.find('{http://www.hmdb.ca}protein_associations')
    if protein_associations is not None: protein_associations = protein_associations.text

    rows.append({
        'name': name,
        'iupac_name': iupac_name,
        'traditional_iupac': traditional_iupac,
        'inchikey' : inchikey,
        'inchi' : inchi,
        'chemical_formula': chemical_formula,
        'average_molecular_weight':average_molecular_weight,
        'description': description,
        'synonyms': synonyms,
        'normal_concentrations': normal_concentrations,
        'abnormal_concentrations': abnormal_concentrations,
        'chemspider_id': chemspider_id,
        'drugbank_id': drugbank_id,
        'foodb_id': foodb_id,
        'pubchem_compound_id': pubchem_compound_id,
        'pdb_id': pdb_id,
        'chebi_id': chebi_id,
        'kegg_id': kegg_id,
        'biocyc_id': biocyc_id,
        'vmh_id': vmh_id,
        'protein_associations': protein_associations
    })
    
df = pd.DataFrame(rows, columns=cols)

df.to_csv('Last_output.csv')

#df.shape is equal to (217920, 21)
```
